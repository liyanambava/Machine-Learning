```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 0)
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
acc = accuracy_score(y_test, lr_pred)
prec = precision_score(y_test, lr_pred)
rec = recall_score(y_test, lr_pred)
f1 = f1_score(y_test, lr_pred)

result = pd.DataFrame([['Logistic Regression', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
```

```python
result
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=lr, X = x_train, y= y_train, cv=10)

print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```

```python
cm = confusion_matrix(y_test, lr_pred)
```

```python
cm
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0, n_estimators=10, criterion = 'entropy')
rf.fit(x_train, y_train)
```

```python
rf_pred = rf.predict(x_test)
```

```python
acc = accuracy_score(y_test, rf_pred)
prec = precision_score(y_test, rf_pred)
rec = recall_score(y_test, rf_pred)
f1 = f1_score(y_test, rf_pred)

result1 = pd.DataFrame([['Random Forest', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
```

```python
result = result.append(result1, ignore_index=True)
```

```python
result
```
